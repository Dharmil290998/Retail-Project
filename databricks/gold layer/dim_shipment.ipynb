```python
shipment = spark.table("sales.silver.shipping")

dim_shipment = shipment.select(
    "shipment_id",
    "carrier",
    "delivery_date",
    "order_id",
    "shipment_date",
    "shipment_status",
    "delivery_days",
    "is_delivered",
    "business_status"
)

dim_shipment.write.mode("append").saveAsTable("sales.gold.dim_shipment")
```
